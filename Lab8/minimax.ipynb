## Câu 2. Code minimax làm sao cho không giống với slide và sau đó nhận xét 2 kết quả

```python
## code minimax không giống slide

import tkinter as tk
from tkinter import messagebox
from functools import partial

# --- Cấu hình mặc định -----------------------------------------------
MARKS = ('X', 'O')   # Ký hiệu hai người chơi
BTN_W, BTN_H = 6, 3  # Kích thước button
WIN_BG = '#80ffaa'   # Màu nền khi thắng

# --- Lớp trò chơi -----------------------------------------------------
class TicTacToe:
    def __init__(self, size: int):
        self.size = size
        self.root = tk.Tk()
        self.root.title(f"{size}×{size} Tic‑Tac‑Toe")
        self.turn = 0          # 0 → X, 1 → O
        self.moves = 0
        self.buttons = []      # Ma trận Button
        self._build_board()
        self.root.mainloop()

    # ---------------------------------------------------------- UI ----
    def _build_board(self):
        for r in range(self.size):
            row_btns = []
            for c in range(self.size):
                b = tk.Button(
                    self.root, text=' ', width=BTN_W, height=BTN_H,
                    font=('Helvetica', 20, 'bold'),
                    command=partial(self._on_click, r, c)
                )
                b.grid(row=r, column=c, sticky='nsew')
                row_btns.append(b)
            self.buttons.append(row_btns)
            self.root.grid_rowconfigure(r, weight=1)
            self.root.grid_columnconfigure(r, weight=1)

    # ---------------------------------------------------- Xử lý click ----
    def _on_click(self, r: int, c: int):
        btn = self.buttons[r][c]
        if btn['text'] != ' ':
            return
        mark = MARKS[self.turn]
        btn.config(text=mark)
        self.moves += 1

        if self._is_winning_move(r, c, mark):
            self._highlight_win(r, c, mark)
            messagebox.showinfo("🎉 Chiến thắng", f"Người chơi {self.turn+1} ({mark}) thắng!")
            self._lock_board()
        elif self.moves == self.size ** 2:
            messagebox.showinfo("🤝 Hoà", "Trò chơi kết thúc với kết quả hoà.")
            self._lock_board()
        else:
            self.turn ^= 1  # Đổi lượt

    # ------------------------------------------- Kiểm tra chiến thắng ----
    def _is_winning_move(self, r: int, c: int, mark: str) -> bool:
        n = self.size
        row = all(self.buttons[r][j]['text'] == mark for j in range(n))
        col = all(self.buttons[i][c]['text'] == mark for i in range(n))
        diag = r == c and all(self.buttons[i][i]['text'] == mark for i in range(n))
        anti = r + c == n - 1 and all(self.buttons[i][n-1-i]['text'] == mark for i in range(n))
        return row or col or diag or anti

    # ----------------------------------------------- Tô màu ô thắng ----
    def _highlight_win(self, r: int, c: int, mark: str):
        n = self.size
        if all(self.buttons[r][j]['text'] == mark for j in range(n)):
            for j in range(n): self.buttons[r][j].config(bg=WIN_BG)
        if all(self.buttons[i][c]['text'] == mark for i in range(n)):
            for i in range(n): self.buttons[i][c].config(bg=WIN_BG)
        if r == c and all(self.buttons[i][i]['text'] == mark for i in range(n)):
            for i in range(n): self.buttons[i][i].config(bg=WIN_BG)
        if r + c == n - 1 and all(self.buttons[i][n-1-i]['text'] == mark for i in range(n)):
            for i in range(n): self.buttons[i][n-1-i].config(bg=WIN_BG)

    # ------------------------------------------- Khóa bàn cờ sau trận ----
    def _lock_board(self):
        for row in self.buttons:
            for btn in row:
                btn.config(state='disabled')

# --- Mục 1: chọn kích thước ------------------------------------------
def launch_menu():
    menu = tk.Tk()
    menu.title("Chọn kích thước bàn cờ")

    tk.Label(menu, text="Chọn kích thước:", font=('Helvetica', 14, 'bold')).pack(pady=10)

    for size in (4, 5, 6, 8):
        tk.Button(
            menu, text=f"{size} × {size}", width=10, height=2,
            font=('Helvetica', 12),
            command=lambda s=size: (menu.destroy(), TicTacToe(s))
        ).pack(pady=5)

    menu.mainloop()

# --- Mục 4: chạy chương trình ----------------------------------------
if __name__ == "__main__":
    launch_menu()
```
